```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset
df = pd.read_csv('movies.csv')
df
```

```python
df.isnull().sum()
```

```python
# --- 3️⃣ Line Plot: Year vs Average Gross ---
avg_gross = df.groupby('year')['gross'].mean()
plt.figure(figsize=(8,5))
plt.plot(avg_gross.index, avg_gross.values, marker='o', color='orange')
plt.title('Average Gross Revenue by Year')
plt.xlabel('Year')
plt.ylabel('Average Gross')
plt.grid(True)
plt.show()
```

```python
# --- 5️⃣ Boxplot: Runtime per Genre ---
plt.figure(figsize=(8,5))
df.boxplot(column='runtime', by='genre', grid=False, patch_artist=True)
plt.title('Runtime Distribution by Genre')
plt.suptitle('')
plt.xlabel('Genre')
plt.ylabel('Runtime (minutes)')
plt.xticks(rotation=45)
plt.show()
```

```python
# --- 6️⃣ Scatter Plot: Score vs Votes ---
plt.figure(figsize=(7,5))
plt.scatter(df['votes'], df['score'], color='red')
plt.title('Votes vs IMDb Score')
plt.xlabel('Votes')
plt.ylabel('IMDb Score')
plt.grid(True)
plt.show()
```
